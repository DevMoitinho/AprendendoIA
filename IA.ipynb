```python
import matplotlib.pyplot as plt
import random
import math

from sklearn import datasets


def sigmoid(x):
    return 1 / (1 + math.exp(-x))

# Load the digits dataset
digits = datasets.load_digits()
```

Modelo


```python
x = (digits.data)[0:50]
w = [[[random.random() for i in range(32)] for e in range(64)],
     [[random.random() for i in range(16)] for e in range(32)],
     [[random.random() for i in range(10)] for e in range(16)]]

def camada(camada, xi,wi):
  z = [[] for i in range(len(wi[0]))]
  for i in range(len(z)):
    for j in range(len(xi)):
      z[i].append(wi[j][i]*xi[j])

  out = []
  for e in z:
    out.append(sigmoid(sum(e)))

  return out
```

FrontFeeding

```python
def sm(a):
  print("-"*30)
  for e in a: print(e)
  print("-"*30)

x_targets = digits.target[0:50]

def getTargets(n):
  t = [0 for i in range(10)]
  t[x_targets[0]] = 1
  return t

def ff(entrada, pesos):
  for i in range(len(pesos)):
    if i == 0:
      a = camada(i,entrada,pesos[i])

    else:
      a = camada(i,a,pesos[i])

  return a

n = 0
teste = x[0]
t = getTargets(n)

y = ff(teste,w)
sm(y)
print(t)

error = sum([((t[i]-y[i])**2) for i in range(10)]) / 2

print(error)
```
